```python
labels = europkillers
subtitles = thoses_countries

#explodelist = (0, 0, 0, 0.1, 0.1, 0.1, 0.2, 0.3)

dico={"color":'gold', "weight":'bold', "fontsize":18}

#color_of_text = "pink"
color_of_text = "aquamarine"

fig, axes = plt.subplots(4, 2, figsize=(18, 22))
fig.suptitle('Causes of Death in Different Countries', fontsize=16)

fig.patch.set_facecolor('lightgrey')

for idx, (cause, deaths) in enumerate(df_for_pie_to_dict.items()):
    ax = axes[idx // 2, idx % 2]
    
    raw_deathvals = list(deaths.values())
    raw_deathkeys = list(deaths.keys())
    
    clean_colors = []
    
    clean_deathvals = []
    clean_keys = []
    
    for val in raw_deathvals:
        if val > 5.0:
            #validx = val.index
            validx = raw_deathvals.index(val)
            
            clean_deathvals.append(val)
            clean_keys.append(raw_deathkeys[validx])
            clean_colors.append(colors_causes[validx])    
        
    #wedges, texts, autotexts = ax.pie(deaths.values(), labels=deaths.keys(),colors=colors, autopct='%1.1f%%', startangle=90,textprops=dico,explode=explodelist)
    wedges, texts, autotexts = ax.pie(clean_deathvals, labels=clean_keys, colors=clean_colors, autopct='%1.1f%%', startangle=90,textprops=dico)
    
    #print(f" wedges {wedges}    texts {texts}     autotexts  {autotexts}     deaths.values()  {deaths.values()}     deaths.keys()   {deaths.keys()}  ")

    for text in texts + autotexts:
        text.set_color(color_of_text)        
        
    ax.set_title(cause)
    
legend = fig.legend(labels, title='Causes', loc='upper center', bbox_to_anchor=(0.5, 0.95), fancybox=True, shadow=True, ncol=3, handlelength=3, borderpad=1.1, labelspacing=1.1, fontsize=12)

legend.get_title().set_color('blue')

plt.tight_layout(rect=[0, 0.03, 1, 0.92])
plt.show()    
```

```python
bbox_props=dict(boxstyle='square,pad=0.3',fc ='w',ec='k',lw=0.72)
kw=dict(xycoords='data',textcoords='data',arrowprops=dict(arrowstyle='-'),zorder=0,va='center')

fig1,ax1=plt.subplots(figsize=(18, 22))
fig1.suptitle('Causes of Death in Nigeria', fontsize=16)

fig1.patch.set_facecolor('lightgrey')

labels = national_diseases_percent_colnames

values = nigeria_pc_valflatlist

annotate_dict = {k:v for k,v in zip(labels, values)}
val = [[x,y] for x,y in zip(sorted(values, reverse=True),sorted(values))]
values1 = sum(val, [])

new_labels = []
for v in values1[:len(values)]:
    for key, value in annotate_dict.items():
        if v == value:
            new_labels.append(key)
            
wedges,texts=ax1.pie(values1[:len(values)], labeldistance=1.2,startangle=90)

for i,p in enumerate(wedges):
    ang=(p.theta2-p.theta1)/2. +p.theta1
    y=np.sin(np.deg2rad(ang))
    x=np.cos(np.deg2rad(ang))
    horizontalalignment={-1:"right",1:"left"}[int(np.sign(x))]
    connectionstyle="angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle":connectionstyle})
    ax1.annotate(new_labels[i],xy=(x, y),xytext=(1.35*np.sign(x),1.4*y),
                horizontalalignment=horizontalalignment,**kw)
    
plt.show()
```
