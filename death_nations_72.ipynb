```python
df.loc[df.index[-1], 'Country Name'] = '__World__'
df
```

```python
columns_to_calculate =df.columns[1:]

for column in columns_to_calculate:
        
    df[f'{column} %'] = (df[column] / df['__All causes__']) * 100

df.drop(columns=["__All causes__ %"], axis=1, inplace=True)
```

```python
df
```

```python
last_row = df.iloc[-1]
#print(last_row)
for column in columns_to_calculate:    
    if not column.endswith("%"):        
        #print("column ", column)
        df[f'{column}_world_pc'] = (df[f"{column}"] / last_row[f"{column}"]) * 100
```

```python
national_diseases_percent_colnames = []
world_diseases_percent_colnames = []


for patho in pathologies:
    national_patho_percent_string = patho + " %"
    world_patho_percent_string = patho + "_world_pc"
    national_diseases_percent_colnames.append(national_patho_percent_string)
    world_diseases_percent_colnames.append(world_patho_percent_string)
```

```python
df_with_percent = df["Country Name"].copy()

colnames = []

for nationalvalue_nationalpercent_worldpercent in list(zip(pathologies, national_diseases_percent_colnames,world_diseases_percent_colnames)):
    
    colnames.append(nationalvalue_nationalpercent_worldpercent[0])
    colnames.append(nationalvalue_nationalpercent_worldpercent[1])
    colnames.append(nationalvalue_nationalpercent_worldpercent[2])

colnames.append("__All causes__") # raw value
#colnames.append("__All causes__ %") # always = 100% of national death ...
colnames.append("__All causes___world_pc") # international 'contribution'

print(colnames)  
```

```python
new_df_T = myselection[europkillers].T
transpo_dico = new_df_T.to_dict()
#transpo_dico
```

```python
labels = europkillers
subtitles = thoses_countries


# Create a subplot (3, 2) for pie charts
fig, axes = plt.subplots(3, 2, figsize=(18, 22))
fig.suptitle('Causes of Death in Different Countries', fontsize=16)

# Set background color to grey
fig.patch.set_facecolor('lightgrey')

# Plot pie charts for each cause of death
for idx, (cause, deaths) in enumerate(transpo_dico.items()):
    ax = axes[idx // 2, idx % 2]
    
    #wedges, texts, autotexts = ax.pie(deaths.values(), labels=deaths.keys(), colors=colors, autopct='%1.1f%%', startangle=90,textprops=dict(color='k', weight='bold', fontsize=8))
    
    wedges, texts, autotexts = ax.pie(deaths.values(), labels=deaths.keys(),colors=colors, autopct='%1.1f%%', startangle=90,textprops=dict(fontsize=15))
    #wedges, texts, autotexts = ax.pie(deaths.values(), labels=europkillers,colors=colors, autopct='%1.1f%%', startangle=90,textprops=dict(fontsize=15))
    
    
    
    

    # Set label color to pink
    for text in texts + autotexts:
        text.set_color('pink')

    ax.set_title(cause)
    
    #ax.legend(fontsize=25)

# Add subtitles for each row
for ax, subtitle in zip(axes[:, 0], subtitles):
    ax.set_ylabel(subtitle, fontsize=20, color='purple')

# Add a single legend at upper center
legend = fig.legend(labels, title='Causes', loc='upper center', bbox_to_anchor=(0.5, 0.95), fancybox=True, shadow=True, ncol=3)
legend.get_title().set_color('pink')

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.92])

# Show the plot
plt.show()
```
